```python
import sys
sys.path.append('..')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from src.data_loader import load_raw
from src.utils import set_style, print_section

set_style()
df = load_raw()
print(f'Shape : {df.shape}')
df.head()
```

## Distribution des prix et revenus

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
df['Price_USD'].hist(bins=30, ax=axes[0], color='steelblue', edgecolor='black')
axes[0].set_title('Distribution des prix'); axes[0].set_xlabel('Prix ($)')

df['Units_Sold'].hist(bins=20, ax=axes[1], color='orange', edgecolor='black')
axes[1].set_title('Distribution des unités vendues')

df['Revenue_USD'].hist(bins=30, ax=axes[2], color='#2ecc71', edgecolor='black')
axes[2].set_title('Distribution des revenus'); axes[2].set_xlabel('Revenue ($)')
plt.tight_layout(); plt.show()
```

## Vérification cohérence Revenue = Price × Units

```python
df['check'] = (df['Revenue_USD'] - df['Price_USD'] * df['Units_Sold']).round(2)
print(f"Écart max : {df['check'].abs().max():.2f}")
print(f"Écart moyen : {df['check'].abs().mean():.4f}")
# Si tout est OK : écart proche de 0 (juste des arrondis flottants)
```
