## Bonus : Boxplot des prix par marque

```python
import seaborn as sns
fig, ax = plt.subplots(figsize=(11, 6))
sns.boxplot(data=df, x='Brand', y='Price_USD',
            order=df.groupby('Brand')['Price_USD'].median().sort_values().index,
            hue='Brand', palette='Set2', legend=False, ax=ax)
ax.set_title('Distribution des prix par marque')
plt.show()
```
